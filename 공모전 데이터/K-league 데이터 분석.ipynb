```python
def format_time(x):
    minutes = int(x // 60)
    seconds_total = x % 60 

    sec_int = int(seconds_total)                # 초 정수부
    sec_frac = seconds_total - sec_int          # 소수부

    # 초 정수부: 항상 2자리
    sec_int_str = f"{sec_int:02d}"

    # 초 소수부: 2자리로 고정
    sec_frac_str = f"{sec_frac:.2f}"[1:]        # '0.xxx' → '.xxx'

    return f"{minutes:02d}:{sec_int_str}{sec_frac_str}"
```

```python
goal_x = 105
goal_y = 34

shots['shot_meter'] = np.sqrt(
    (goal_x - shots['adj_x'])**2 +
    (goal_y - shots['adj_y'])**2   
)
```

```python
goal_x = 105
goal_y = 34
goal_y_half = 3.66

left_post = goal_y - goal_y_half
right_post = goal_y + goal_y_half
```

```python
shots['shot_angle'] = np.abs(
        np.arctan2(right_post - shots['adj_y'], goal_x - shots['adj_x']) -
        np.arctan2(left_post - shots['adj_y'], goal_x - shots['adj_x'])
)
```

```python
#우선 경기장을 분할 한다(12 x 6칸으로 분할)
xt_grid = np.array([
 [0.000,0.001,0.002,0.003,0.004,0.003,0.002,0.001],
 [0.001,0.002,0.004,0.006,0.008,0.006,0.004,0.002],
 [0.002,0.004,0.008,0.012,0.016,0.012,0.008,0.004],
 [0.003,0.006,0.012,0.020,0.030,0.020,0.012,0.006],
 [0.004,0.008,0.016,0.030,0.050,0.030,0.016,0.008],
 [0.006,0.012,0.020,0.040,0.070,0.040,0.020,0.012],
 [0.008,0.016,0.030,0.060,0.100,0.060,0.030,0.016],
 [0.010,0.020,0.040,0.080,0.150,0.080,0.040,0.020],
 [0.012,0.025,0.050,0.100,0.200,0.100,0.050,0.025],
 [0.015,0.030,0.060,0.120,0.250,0.120,0.060,0.030],
 [0.020,0.040,0.080,0.160,0.350,0.160,0.080,0.040],
 [0.030,0.060,0.120,0.250,0.500,0.250,0.120,0.060]
])
```

```python
# 마찬가지 홈 원정 좌표가 반대로 나오다 보니 공격 방향 좌표 통일 
# home_team 공격 기준으로 원정팀의 값을 좌우 반전 시키는 코드
pitch_length = 105
pitch_width = 68


df = soccer_df.copy()
df['adj_x'] = np.where(df['team_id'] == df['home_team_id'],
                      df['start_x'],
                      pitch_length - df['start_x']) #원정팀 좌우 반전

df['adj_y'] = np.where(df['team_id'] == df['home_team_id'],
                      df['start_y'],
                      pitch_width - df['start_y'])

df['adj_end_x'] = np.where(df['team_id'] == df['home_team_id'],
                      df['end_x'],
                      pitch_length - df['end_x'])

df['adj_end_y'] = np.where(df['team_id'] == df['home_team_id'],
                      df['end_y'],
                      pitch_width - df['end_y'])
```

```python
# 좌표가 몇번째 칸에 속해 있냐 

def get_xt(x, y, grid):
    x_bin = min(int(x / 105 * grid.shape[0]), grid.shape[0] - 1)
    y_bin = min(int(y / 68 * grid.shape[1]), grid.shape[1] - 1)
    return grid[x_bin, y_bin]
```

```python
# 행동 전/후 위협도 계산

df['xt_start'] = df.apply(
    lambda r: get_xt(r['adj_x'], r['adj_y'], xt_grid),
    axis=1
)

df['xt_end'] = df.apply(
    lambda r: get_xt(r['adj_end_x'], r['adj_end_y'], xt_grid),
    axis=1
)
```

```python
# xt값 최종 계산

df['xt'] = df['xt_end'] - df['xt_start']
```

```python
# 슛은 0으로 처리 xg값이 있기 때문
df.loc[df['type_name'] == 'Shot', 'xt'] = 0
```

```python
# 누적 xg값을 그대로 쓰지말고 10분 단위로 변화하는 누적 xg값을 만든다 

def last_10min_xg(g):
    g = g.copy()
    g['last_10min_xgdiff'] = 0.0
    
    for i in range(len(g)):
        t = g.iloc[i]['time_seconds']
        past = g[g['time_seconds'] <= t - 600]
        
        if len(past) > 0:
            g.iloc[i, g.columns.get_loc('last_10min_xgdiff')] = (
            g.iloc[i]['xg_diff'] - past.iloc[-1]['xg_diff'])
        
        else:
            g.iloc[i, g.columns.get_loc('last_10min_xgdiff')] = (g.iloc[i]['xg_diff'])
        
    return g

# 경기의 흐름을 전반적으로 나타내면서 어느팀이 더 위협적인지를 나타내는 지표
```

```python
soccer_df = soccer_df.groupby('game_id', group_keys = False).apply(last_10min_xg)
```

```python
soccer_df['remain_time'] = 90*60 - soccer_df['time_seconds']
```

```python
feature = ['home_scorediff', 'home_xgdiff', 'home_xadiff',
           'home_xtdiff', 'time_seconds', 'remain_time']

x = soccer_df[feature]
y = soccer_df['game_result']
```

```python
# 로지스틱 회귀로 실시간 승률을 구해보자
# trina, test 분리

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                   test_size=0.2,
                                                   random_state=30,
                                                   stratify=y)
                                                
```

```python
# 스케일링 하기

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.fit_transform(x_test)
```

```python
# 다중 로지스틱 회귀 학습

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class = 'multinomial',
                           max_iter=1000,
                           solver='lbfgs')

model.fit(x_train_scale, y_train)
```

```python
proba = model.predict_proba(scaler.transform(x))
```

```python
proba_df = pd.DataFrame(proba, columns=model.classes_,
                       index= x.index)
```

```python
soccer_df.loc[x.index, 'draw_prob']     = proba_df[0]
soccer_df.loc[x.index, 'home_win_prob'] = proba_df[1]
soccer_df.loc[x.index, 'away_win_prob'] = proba_df[2]
```
