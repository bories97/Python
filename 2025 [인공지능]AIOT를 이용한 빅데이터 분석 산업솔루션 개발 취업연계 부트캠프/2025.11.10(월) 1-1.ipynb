```python
# stack 함수
```

```python
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz',
                     'foo', 'foo', 'qux', 'qux'],
                    ['one', 'two', 'one', 'two',
                     'one', 'two', 'one', 'two']]))
```

```python
tuples #zip 함수는 어 각 리스트에서 순서대로 값들을 묶어서 출력한다.
```

```python
index = pd.MultiIndex.from_tuples(tuples, name = ['F', 'S'])
```

```python
index
```

```python
df4 = pd.DataFrame(np.random.randn(8, 2), index=index, columns=['A', 'B'])
```

```python
df4
```

```python
df4.stack()
```

```python
df5 = pd.DataFrame({'A' : ['one', 'one', 'two', 'three'] * 3,
                   'B' : ['A', 'B', 'C'] * 4,
                   'C' : ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 2,
                   'D' : np.random.randn(12),
                   'E' : np.random.randn(12)})
```

```python
df5
```

```python
pd.pivot_table(df5, values='D', index=['A', 'B'], columns = ['C'])
```

```python
# 그래프
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
```

```python
ts1 = ts.cumsum()
```

```python
plt.ylabel('num')
plt.xlabel('year')
ts1.plot()
```

```python
df6 = pd.DataFrame(np.random.randn(1000, 4), index=ts.index,
                  columns=['A', 'B', 'C', 'D'])
```

```python
df6
```

```python
df6 = df6.cumsum()
```

```python
plt.figure()
df6.plot()
plt.legend(loc = 'best')
plt.show()
```

```python
plt.figure(); df6.plot(); plt.legend(loc='best')
```
