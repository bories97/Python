```python
def x(a, b):
    count = 0
    for name in a:
        if name.startswith(b):
            count = count + 1
    return count
```

```python
print("'이' 씨는", x(data1, "이"), "명입니다.")
print("'김' 씨는", x(data1, "김"), "명입니다.")
print("'박' 씨는", x(data1, "박"), "명입니다.")
```
