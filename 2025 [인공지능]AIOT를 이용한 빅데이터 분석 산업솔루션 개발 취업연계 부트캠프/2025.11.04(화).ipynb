```python
def gugudan(n):
    result = []
    i = 1
    while i <= 9:
        result.append(n*i)
        i += 1
    return result
```

```python
print(gugudan(3))
```

```python
gugu_reslult = []
for i in range(2, 10):
    print(i)
    gugu_reslult.append(gugudan(i))
```
