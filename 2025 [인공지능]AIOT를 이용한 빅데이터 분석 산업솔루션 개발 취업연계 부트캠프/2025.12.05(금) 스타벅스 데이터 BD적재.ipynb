```python
starbucks_info = []

for i in starbucks_list:
    name = i.select('strong')[0].text.strip()
    lat = i['data-lat'].strip()
    long = i['data-long'].strip()
    store_type = i.select('i')[0]['class'][0][4:]
    address = str(i.select("#mCSB_3 p.result_details")[0]).split('<br/>')[0].split('>')[1]
    tel = str(i.select("#mCSB_3 p.result_details")[0]).split('<br/>')[1].split('</p>')[0]
    
    starbucks_info.append([name, lat, long, store_type, address, tel])
```

```python
starbucks_info
```

```python
starbucks_info_df = pd.DataFrame(starbucks_info, columns=['이름', '위도', 
                                                         '경도', '종류',
                                                          '주소', '전화번호'])
starbucks_info_df
```


### DB적재

```python
import pymysql
from sqlalchemy import create_engine
```

```python
conn = pymysql.connect(host='127.0.0.1',port = 3306, user='root', 
                       password='1234', db='dbmaria', charset='utf8')
```

```python
cur = conn.cursor()
```

```python
cur.execute(''' CREATE TABLE starbucks_info (
                id INT AUTO_INCREMENT PRIMARY KEY,
                이름 VARCHAR(100),
                위도 FLOAT,
                경도 FLOAT,
                종류 VARCHAR(50),
                주소 VARCHAR(200),
                전화번호 VARCHAR(20)
                )''')
```

```python
for index, row in starbucks_info_df.iterrows():
    cur.execute("""
        INSERT INTO starbucks_info (
        이름, 위도, 경도, 종류, 주소, 전화번호)
        VALUES (%s, %s, %s, %s, %s, %s)""",
               (
               row['이름'],
               row['위도'],
               row['경도'],
               row['종류'],
               row['주소'],
               row['전화번호']
               ))
```
