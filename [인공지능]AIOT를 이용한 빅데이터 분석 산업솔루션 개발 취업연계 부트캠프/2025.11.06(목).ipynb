```python
def guess(n):
    #n = int(input("숫자 입력")) <- 함수로 지정해서 필요 x
    result = []
    while n != 1:
        result.append(n)
        if n % 2 == 0:
            n = int(n / 2)
        else:
            n = 3*n + 1
        if n == 1:
            print(1)
            result.append(1)
    return result
```

```python
def cycle(n):
    if n == 1:  # n = 1일때
        return [1]
    if n % 2 == 0:  #n = 짝수일때
        return [n] + cycle(n//2)
    else:         #n = 홀수일때
        return [n] + cycle(3*n + 1)
```

```python
cycle(9)
```

```python
def quez(a):
    result = []
    while a != 1:
        result.append(a)
        if a % 2 == 0:
            a = int(a/2)
        else:
            a = a*3 + 1
        if a == 1:
            result.append(1)
    return result
```
