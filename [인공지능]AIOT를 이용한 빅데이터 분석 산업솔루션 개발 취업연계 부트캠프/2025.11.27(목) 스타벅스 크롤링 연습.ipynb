```python
starbucks_list = []

for item in starbucks_soup_list:
    name = item.select('strong')[0].text.strip()
    lat = item['data-lat'].strip()
    lng = item['data-long'].strip()
    store_type = item.select('i')[0]['class'][0][4:]
    address = str(item.select('#mCSB_3 p.result_details')[0]).split('<br/>')[0].split('>')[1]
    tel = str(item.select('#mCSB_3 p.result_details')[0]).split('<br/>')[1].split('<')[0]

    starbucks_list.append([name, lat, lng, store_type, address, tel])

starbucks_list
```

```python
# starbucks_list : [[name, lat, lng, store_type, address, tel], ...] 형태
starbucks_df_list = []

for item in starbucks_list:
    name = item[0]
    lat = float(item[1])   # 문자열 → 실수
    lng = float(item[2])
    address = item[4]
    gu = address.split(" ")[1]  # 주소에서 구 추출
    brand = "스타벅스"

    starbucks_df_list.append({
        "매장이름": name,
        "주소": address,
        "구": gu,
        "브랜드": brand,
        "위도": lat,
        "경도": lng
    })

# DataFrame 생성
df = pd.DataFrame(starbucks_df_list)
df
```

```python
import folium

# 1) 지도 중심 설정 (한국 평균 좌표)
map_center = [df['위도'].mean(), df['경도'].mean()]
starbucks_map = folium.Map(location=map_center, zoom_start=9)

# 2) 매장별 마커 추가
for idx, row in df.iterrows():
    folium.Marker(
        location=[row['위도'], row['경도']],
        popup=f"{row['매장이름']}<br>{row['주소']}",
        icon=folium.Icon(color='green', icon='coffee', prefix='fa')
    ).add_to(starbucks_map)

# 3) 지도 표시
starbucks_map
```

```python
url = "http://books.toscrape.com/"
res = requests.get(url)
html = res.text
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
books_list = []

for book in soup.select("article.product_pod"):
    title = book.h3.a["title"]  # 책 제목
    price = book.select_one("p.price_color").text  # 가격
    rating = book.p["class"][1]  # 평점 (One, Two, Three, Four, Five)
    books_list.append([title, price, rating])
```

```python
df_books = pd.DataFrame(books_list, columns=["제목", "가격", "평점"])
df_books
```
