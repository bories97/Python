```python
coffee = 10      # 남은 커피 개수
price = 1500    # 커피 한 잔 가격

while True:
    money = int(input("돈을 넣어주세요: "))
    if money == price:
        print("금액 지불이 완료되었습니다. 커피를 드립니다.")
        coffee = coffee - 1
    elif money > price:
        print("거스름돈은 %d원 입니다." % (money - price))
        print("커피를 드립니다.")
        coffee = coffee - 1
        print("남은 커피는 %d개 입니다." % coffee)
    else:
        print("돈이 부족합니다. 다시 지불해주세요.")
    if coffee == 0:
        print("오늘 준비된 재고가 다 소진되었습니다.")
        break
```

### 중첩된 while 문(nested)

```python
#구구단 만들기 i j 
i = 1

while i < 10:
    print("2 *", i, "=", 2*i)
    i += 1

     

```

```python
e = 2
while e <= 3:
    i = 1
    while i <= 9:
        print(e, "x", i, "=", e*i)
        i += 1
    print()
    e += 1
```

```python
i = 1
while i <= 4:
    j = 1
    while j <= 9:       
        print(f"i = {i}, j = {i*j}")
        j += 1
    i += 1
```

```python
i = 2
while i <= 9:
    j = 1
    while j <= 9:
        print(f"{i}", "X", f"{j}", "=", i*j, end = "  ")
        j += 1
    print()
    i += 1
```

```python
i = 2
result = ""
while i <= 9:
    j = 1
    while j <= 9:
        result += f"{i} X {j} = {i*j}\t"
        j += 1
    result += "\n"
    i += 1
    
print(result)
```

```python
for j in range(1, 10):
    for i in range(2, 10):
        print(f"{i} X {j} = {i*j}", end = "\t")
    print()
```
