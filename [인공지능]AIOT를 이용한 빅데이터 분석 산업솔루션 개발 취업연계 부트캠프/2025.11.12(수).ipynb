```python
# 2020년 4월 버전의 위키피디아 미국 ETF 웹 페이지
url2 = "https://en.wikipedia.org/w/index.php?title=List_of_American_exchange-traded_funds&oldid=948664741"

# User-Agent 헤더 추가 (브라우저처럼 보이게)
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# 웹 페이지에서 정보 스크래핑
print("웹 페이지 다운로드 중...")
resp = requests.get(url2, headers=headers)
print(f"응답 상태 코드: {resp.status_code}")
print(f"응답 길이: {len(resp.text)} 문자\n")

# BeautifulSoup으로 파싱
soup = BeautifulSoup(resp.text, 'lxml')

# 본문 내용 찾기
content = soup.find('div', {'id': 'mw-content-text'})

if content is None:
    content = soup.find('div', {'class': 'mw-parser-output'})

if content is None:
    print("⚠️ 본문을 찾을 수 없습니다.")
    content = soup

print(f"본문 찾기 완료\n")

etfs = {}
current_category = "Unknown"
current_subcategory = None

# 본문의 모든 요소를 순회
for element in content.find_all(['h2', 'h3', 'h4', 'li']):
    
    # 카테고리 헤더 찾기 (h2, h3, h4)
    if element.name in ['h2', 'h3', 'h4']:
        header_text = element.get_text().strip()
        
        # 편집 링크나 불필요한 텍스트 제거
        header_text = re.sub(r'\[edit\]', '', header_text).strip()
        
        # 헤더가 비어있지 않고 의미있는 카테고리인 경우
        if header_text and len(header_text) > 0:
            # h2는 대분류, h3/h4는 소분류로 처리
            if element.name == 'h2':
                current_category = header_text
                current_subcategory = None
                print(f"대분류: {current_category}")
            elif element.name in ['h3', 'h4']:
                current_subcategory = header_text
                print(f"  소분류: {current_subcategory}")
    
    # ETF 정보 추출 (li 태그)
    elif element.name == 'li':
        try:
            text = element.get_text().strip()
            
            # 다양한 패턴 매칭
            patterns = [
                # 패턴 1: Name (NYSE Arca|TICKER)
                r'(.+?)\s*\(\s*NYSE Arca\s*\|\s*([A-Z]+)\s*\)',
                # 패턴 2: Name (NASDAQ|TICKER)
                r'(.+?)\s*\(\s*NASDAQ\s*\|\s*([A-Z]+)\s*\)',
                # 패턴 3: Name (NYSE|TICKER)
                r'(.+?)\s*\(\s*NYSE\s*\|\s*([A-Z]+)\s*\)',
                # 패턴 4: Name (Cboe BZX|TICKER)
                r'(.+?)\s*\(\s*Cboe BZX\s*\|\s*([A-Z]+)\s*\)',
                # 패턴 5: 일반 (MARKET|TICKER)
                r'(.+?)\s*\(\s*([^|()]+?)\s*\|\s*([A-Z]{1,5})\s*\)',
            ]
            
            for pattern in patterns:
                match = re.search(pattern, text)
                if match:
                    groups = match.groups()
                    
                    if len(groups) == 2:
                        etf_name = groups[0].strip()
                        etf_ticker = groups[1].strip()
                        
                        # 시장 정보 추출
                        if 'NYSE Arca' in text:
                            etf_market = 'NYSE Arca'
                        elif 'NASDAQ' in text:
                            etf_market = 'NASDAQ'
                        elif 'Cboe BZX' in text:
                            etf_market = 'Cboe BZX'
                        elif 'NYSE' in text:
                            etf_market = 'NYSE'
                        else:
                            etf_market = 'Unknown'
                        
                        # 티커가 유효한지 확인 (1-5자 대문자)
                        if len(etf_ticker) >= 1 and len(etf_ticker) <= 5:
                            # 카테고리 정보 포함
                            full_category = current_category
                            if current_subcategory:
                                full_category = f"{current_category} > {current_subcategory}"
                            
                            etfs[etf_ticker] = [etf_market, etf_name, current_category, current_subcategory if current_subcategory else '']
                        break
                        
                    elif len(groups) == 3:
                        etf_name = groups[0].strip()
                        etf_market = groups[1].strip()
                        etf_ticker = groups[2].strip()
                        
                        # 티커가 유효한지 확인
                        if len(etf_ticker) >= 1 and len(etf_ticker) <= 5:
                            # 카테고리 정보 포함
                            full_category = current_category
                            if current_subcategory:
                                full_category = f"{current_category} > {current_subcategory}"
                            
                            etfs[etf_ticker] = [etf_market, etf_name, current_category, current_subcategory if current_subcategory else '']
                        break
                
        except Exception as err:
            continue

# 결과 확인
print(f"\n{'='*80}")
print(f"총 {len(etfs)}개의 ETF를 찾았습니다.")
print(f"{'='*80}\n")

if len(etfs) > 0:
    # 카테고리별 개수 확인
    categories = {}
    for ticker, info in etfs.items():
        cat = info[2]
        categories[cat] = categories.get(cat, 0) + 1
    
    print("카테고리별 ETF 개수:")
    for cat, count in sorted(categories.items()):
        print(f"  {cat}: {count}개")
    
    print(f"\n추출된 ETF 샘플 (처음 20개):")
    for i, (ticker, info) in enumerate(list(etfs.items())[:20]):
        category_str = info[2]
        if info[3]:
            category_str = f"{info[2]} > {info[3]}"
        print(f"  {ticker:5s} | {info[0]:15s} | {category_str:40s} | {info[1][:40]}")
    
    # etfs 딕셔너리를 데이터프레임으로 변환
    df = pd.DataFrame(etfs).T
    df.columns = ['Market', 'ETF Name', 'Category', 'Subcategory']
    df.index.name = 'Ticker'
    
    print(f"\n데이터프레임 정보:")
    print(f"  - 크기: {df.shape[0]}행 x {df.shape[1]}열")
    print(f"  - 인덱스: {df.index.name}")
    print(f"  - 컬럼: {list(df.columns)}")
    
    print("\n데이터프레임 미리보기:")
    print(df.head(20))
    
    # 카테고리별로 정렬
    df_sorted = df.sort_values(['Category', 'Subcategory', 'Ticker'])
    
    # CSV로 저장
    df_sorted.to_csv('etfs_with_categories.csv', encoding='utf-8-sig')
    print("\n✓ 'etfs_with_categories.csv' 파일로 저장되었습니다.")
    
    # 카테고리별 통계
    print("\n카테고리별 통계:")
    print(df.groupby('Category').size().sort_values(ascending=False))
    
else:
    print("⚠️ ETF 데이터를 찾을 수 없습니다.")
    print("\n페이지 내용 샘플 (처음 1000자):")
    print(soup.get_text()[:1000])
```

```python
import time
from datetime import datetime, timedelta
```

```python
code = "005930"   # 삼성전자
base_url = "https://finance.naver.com/item/sise_day.nhn?code={code}&page={page}"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0 Safari/537.36"
}
```

```python
# 목표 시작일 (오늘 기준 5년 전)
end_date = datetime.today()
start_date = end_date - timedelta(days=365*5)

all_dfs = []
```

```python
# 1) 첫 페이지에서 마지막 페이지 번호(전체 페이지 수) 확인
r = requests.get(base_url.format(code=code, page=1), headers=headers)
r.raise_for_status()
soup = BeautifulSoup(r.text, "lxml")
```

```python
# 네이버 패턴: td.pgRR 안 a href에 마지막 페이지가 있음
last_page_tag = soup.select_one("td.pgRR a")
if last_page_tag:
    href = last_page_tag['href']
    # href like "...&page=681" -> split to get page number
    last_page = int(href.split('=')[-1])
else:
    last_page = 1
```

```python
# 2) 페이지 순회해서 표 읽기; 날짜가 start_date 이전이면 중단
for page in range(1, last_page + 1):
    url = base_url.format(code=code, page=page)
    # pandas.read_html로 바로 표 읽어도 되지만, requests+BeautifulSoup로 얻어 안전하게 처리
    r = requests.get(url, headers=headers)
    r.raise_for_status()
    dfs = pd.read_html(r.text, header=0)  # 페이지에 표 여러개일수 있음; 보통 첫번째가 데이터
    if len(dfs) == 0:
        break
    df = dfs[0]
    # 표 정리: 보통 빈행 또는 헤더가 반복되는 행 제거 필요
    df = df.dropna(how="all")
    # 컬럼 영어로 바꾸면 편함(옵션)
    df.columns = [c.strip() for c in df.columns]
    # 날짜 컬럼 형 변환 (네이버는 '날짜' 형식)
    if '날짜' in df.columns:
        df['Date'] = pd.to_datetime(df['날짜'], format="%Y.%m.%d")
    else:
        # 컬럼명이 영어라면 'Date' 혹은 '날짜'를 찾아 바꾸세요
        df['Date'] = pd.to_datetime(df.iloc[:,0])
    # 숫자 컬럼 쉼표 제거 및 숫자형으로
    for col in ['종가','시가','고가','저가','거래량']:
        if col in df.columns:
            df[col] = df[col].astype(str).str.replace(',', '').str.replace('-', '')
            df[col] = pd.to_numeric(df[col], errors='coerce')
    all_dfs.append(df)

    # 중단조건: 이 페이지의 가장 오래된 날짜가 start_date 이전이면 멈추기
    page_min_date = df['Date'].min()
    if page_min_date <= pd.to_datetime(start_date):
        print(f"Stopped at page {page}, earliest date {page_min_date.date()}")
        break

    time.sleep(0.5)  # 요청 속도 제한: 0.5초 쉬기 (사이트 부하 고려)
```

```python
# 합치기 & 정리
big_df = pd.concat(all_dfs, ignore_index=True)
big_df = big_df.drop_duplicates(subset=['Date']).sort_values('Date', ascending=True).reset_index(drop=True)

# 필요한 컬럼만 선택 및 날짜 필터 (정밀)
big_df = big_df[['Date','종가','시가','고가','저가','거래량']]
big_df = big_df[big_df['Date'] >= pd.to_datetime(start_date)].reset_index(drop=True)

print(big_df.tail())
# 저장
big_df.to_csv('samsung_5y_from_naver.csv', index=False)
```
