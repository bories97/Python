```python
# ver.1
number = [90, 25, 67, 45, 80]
pice = 0
for number in number:
    pice += 1
    if number >= 50:
        print("%d번 학생은 합격입니다." % pice)
    else:
        print("%d번 학생은 실격입니다." % pice)
```

```python
# ver.2
number = [90, 25, 67, 45, 80]
pice = 0
for number in number:
    pice += 1
    if number >= 50:
        print(f"{pice}번학생은 합격입니다")
    else:
        print(f"{pice}번학생은 불합격입니다")
```
